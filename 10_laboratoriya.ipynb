```python
import pandas as pd
```

```python
df = pd.read_csv("22.csv")

print("Dataset muvaffaqiyatli yuklandi")
df.head()
```

```python
head_data = df.head()

print(head_data)

head_data.to_csv("head.csv", index=False)
```

```python
tail_data = df.tail()

print(tail_data)

tail_data.to_csv("tail.csv", index=False)
```

```python
describe_data = df.describe()

print(describe_data)

describe_data.to_csv("describe.csv")
```

```python
percentile_data = df.describe(percentiles=[0.3,0.5,0.7])

print(percentile_data)

percentile_data.to_csv("percentiles.csv")
```

```python
include_data = df.describe(include="all")

print(include_data)

include_data.to_csv("include.csv")
```

```python
exclude_data = df.describe(exclude=["object"])

print(exclude_data)

exclude_data.to_csv("exclude.csv")
```

```python
import sys

with open("info.csv", "w") as f:
    df.info(buf=f)
```

```python
transpose_data = df.describe().T

print(transpose_data)

transpose_data.to_csv("transpose.csv")
```
